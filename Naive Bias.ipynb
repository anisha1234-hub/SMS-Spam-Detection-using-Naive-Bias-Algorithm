# NAIVE BAYES CLASSIFIER FOR SPAM DETECTION

#### Import required libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

#### Load the dataset with correct encoding

```python
df = pd.read_csv("spam.csv", encoding="latin-1")
df.head()
```

#### Keep only required columns and rename them

```python
df = df[['v1', 'v2']]
df.columns = ['label', 'message']
```

#### Display dataset preview

```python
print("Dataset Preview:")
print(df.head())
```

#### Convert labels to numerical values : ham -> 0, spam -> 1

```python
df['label'] = df['label'].map({'ham': 0, 'spam': 1})
```

#### Separate features and target

```python
X = df['message']
y = df['label']
```

#### Split dataset into training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

##  TEXT VECTORIZATION

```python
vectorizer = CountVectorizer(stop_words='english')

X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

##  NAIVE BAYES MODEL

```python
model = MultinomialNB()

# Train the model
model.fit(X_train_vec, y_train)

# Make predictions
y_pred = model.predict(X_test_vec)
```

## MODEL EVALUATION

```python
print("\nAccuracy:", accuracy_score(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

## TEST WITH CUSTOM MESSAGES

```python
def predict_message(text):
    text_vector = vectorizer.transform([text])
    prediction = model.predict(text_vector)
    return "Spam" if prediction[0] == 1 else "Ham"

print("\nCustom Predictions:")
print(predict_message("Congratulations! You won a free lottery ticket"))
print(predict_message("Can we meet tomorrow for the project discussion?"))
```
